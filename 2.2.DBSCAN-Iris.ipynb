```python
#import libraries
from sklearn.cluster import DBSCAN 
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from scipy.stats import zscore
import numpy as np
```

```python
#Load Dataset
meas = load_iris().data
X = meas[:,[2,3]]
```

```python
#Implement DBSCAN
dbscan = DBSCAN(eps = .1, min_samples = 5).fit(X)
IDX = dbscan.labels_

#Scatterplot of dimensions [2, 3] of iris dataset
plt.figure(1)
plt.scatter(X[:,0], X[:,1])
plt.title("Scatterplot of dimensions [2, 3] of the iris dataset")
plt.show()

##Plot after clustering##
plt.figure(1)
plt.scatter(X[:,0], X[:,1], c=IDX)
plt.title("Clusters after DBSCAN")
plt.show()
```

```python
#Scale data with zscore
xV1 = zscore(X[:,0])
xV2 = zscore(X[:,1])
X_scaled=np.array(list(zip(xV1,xV2)))

#Implement DBSCAN
dbscan = DBSCAN(eps = 0.1, min_samples = 5).fit(X_scaled)
IDX = dbscan.labels_

#Scatterplot of scaled dimensions [2, 3] of iris dataset
plt.figure(1)
plt.scatter(xV1, xV2)
plt.show()

##Plot after clustering##
plt.figure(1)
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=IDX)
plt.title("Clusters & noise after DBSCAN")
plt.show()
```
