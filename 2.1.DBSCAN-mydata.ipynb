```python
#import libraries
import scipy.io
import numpy as np
from sklearn.cluster import DBSCAN 
import matplotlib.pyplot as plt
```

```python
#Load Datase
mat_file = scipy.io.loadmat('mydata.mat')
X = np.array(mat_file['X'])
```

```python
#Implement DBSCAN
dbscan = DBSCAN(eps = 0.5, min_samples = 15).fit(X)
IDX = dbscan.labels_

#Scatterplot of first 2 dimensions
plt.figure(1)
plt.scatter(X[:,0], X[:,1])
plt.title("Scatterplot of first 2 dimensions")
plt.show()

##Plot after clustering##
plt.figure(1)
plt.scatter(X[:,0], X[:,1], c=IDX)
plt.title("Clusters after DBSCAN")
plt.show()
```
