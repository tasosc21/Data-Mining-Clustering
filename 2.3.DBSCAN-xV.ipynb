```python
#import libraries
import scipy.io
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
#Load Dataset
mat_file = scipy.io.loadmat('xV.mat')
xV = np.array(mat_file['xV'])
```

# Βημα 2

```python
#Dimension Selection
X = xV[:,[0,1]]

#Implement DBSCAN
eps = .3
min_samples = 50
dbscan = DBSCAN(eps = eps, min_samples = min_samples).fit(X)
IDX = dbscan.labels_

#CLustering Plot
plt.figure(1)
plt.scatter(X[:,0], X[:,1], c=IDX)
plt.title("DBSCAN plot of dimensions [0, 1] of the xV dataset")
plt.figtext(.5,.2, f'eps = {eps}, min_samples = {min_samples}')
plt.show()
```
