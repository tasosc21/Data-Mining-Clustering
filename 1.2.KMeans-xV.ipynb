```python
#import libraries
import scipy.io
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from math import dist
from sklearn import metrics
```

```python
#Load dataset
mat_file = scipy.io.loadmat('xV.mat')
xV = np.array(mat_file['xV'])
```

# Βήμα 2

```python
#Selection of first 2 dimensions of the dataset
X = xV[:,[0, 1]]

#Calculate number of rows and columns
numberOfRows, numberOfColumns = X.shape

#Number of clusters
k = 3

#Implementation of k-means
kmeans = KMeans(n_clusters=k).fit(X)
IDX = kmeans.labels_
C = kmeans.cluster_centers_

#SSE calculation
sse = 0.0
for i in range(k):
    for j in range(numberOfRows):
        if IDX[j] == i:
            sse = sse + dist(X[j], C[i])**2
            

#Clustering plot   
plt.plot(X[IDX==0][:,0],X[IDX==0][:,1], 'limegreen', marker='o', linewidth=0, label='C1')
plt.plot(X[IDX==1][:,0],X[IDX==1][:,1], 'yellow', marker='o', linewidth=0, label='C2')
plt.plot(X[IDX==2][:,0],X[IDX==2][:,1], 'c', marker='o', linewidth=0, label='C3')
plt.scatter(C[:,0], C[:,1], marker='x', color='black', s=150, linewidth=3, label='Centroids', zorder=10)
plt.title('Clustering of first 2 dimensions of xV Dataset')
plt.legend()
plt.show()

#Display SSE & SC
print(f'SSE = {sse}')
print(f'SC = {metrics.silhouette_score(X, IDX)}')
```
